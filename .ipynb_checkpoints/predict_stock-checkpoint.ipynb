# Clean data
Bởi vì dữ liệu ban đầu bao gồm rất nhiều giá cổ phiếu trong 1 ngày ở nhiều thời điểm khác nhau. Việc này sẽ gây khó khăn trong quá trình phân tích cũng như xây dựng mô hình. Bên cạnh đó đối với bài toán dạng Time series, các sample theo thời gian nên có khoảng cách tương đối giống nhau giữa các sample. Vì vậy ta sẽ bỏ bớt các sample trong cùng một ngày và giữ lại giá cuối cùng ghi nhận trong ngày. 

```python
for i, stock in enumerate(stock_list):
    stock['Date/Time'] = pd.to_datetime(stock['Date/Time']).dt.date
    stock.drop_duplicates(subset=['Date/Time'], keep='last', inplace=True)
    stock.reset_index(drop=True, inplace=True)
    stock.set_index('Date/Time', inplace=True)
    stock_list[i] = stock
```

# Moving average

```python
ma_day = [3, 7, 14]

for ma in ma_day:
    for stock in stock_list:
        column_name = f"MA for {ma} days"
        mv = stock['Close'].rolling(ma).mean()
        stock[column_name] = mv
       
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

FPT[['Close', 'MA for 3 days', 'MA for 7 days', 'MA for 14 days']].tail(100).plot(ax=axes[0,0])
axes[0,0].set_title('FPT')

MSN[['Close', 'MA for 3 days', 'MA for 7 days', 'MA for 14 days']].tail(100).plot(ax=axes[0,1])
axes[0,1].set_title('MSN')

PNJ[['Close', 'MA for 3 days', 'MA for 7 days', 'MA for 14 days']].tail(100).plot(ax=axes[1,0])
axes[1,0].set_title('PNJ')

VIC[['Close', 'MA for 3 days', 'MA for 7 days', 'MA for 14 days']].tail(100).plot(ax=axes[1,1])
axes[1,1].set_title('VIC')

fig.tight_layout()
```

Chúng ta thấy trong biểu đồ rằng các giá trị tốt nhất để đo Moving average là 3 và 7 ngày vì chúng vẫn nắm bắt trend trong dữ liệu mà không noise. Giá trị 7 ngày cho thấy khả năng nắm bắt trend tốt hơn.

# Building model

```python
plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(FPT['Close'])
plt.xlabel('Date/Time', fontsize=18)
plt.ylabel('Close Price', fontsize=18)
plt.show()
```

```python
FPT.shape
```

```python
# Create a new dataframe with only the 'Close column 
data = FPT.filter(['Close'])
# Convert the dataframe to a numpy array
dataset = data.values
# Get the number of rows to train the model on
training_data_len = int(np.ceil( len(dataset) * .95 ))

training_data_len
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
# Create the training data set 
# Create the scaled training data set
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()
        
# Convert the x_train and y_train to numpy arrays 
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
model.summary()
```

```python
# Create the testing data set
time_step = 15
test_data = scaled_data[training_data_len - time_step: , :]
# Create the data sets x_test and y_test
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(time_step, len(test_data)):
    x_test.append(test_data[i-time_step:i, 0])
    
# Convert the data to a numpy array
x_test = np.array(x_test)

# Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

# Get the models predicted price values 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
# Plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel(None)
plt.ylabel('Close Price', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

# Predicting next 7 days

```python
x_input=test_data[len(test_data)-time_step:].reshape(1,-1)
temp_input=list(x_input)
temp_input=temp_input[0].tolist()

from numpy import array

lst_output=[]
n_steps=time_step
i=0
pred_days = 7 #This is N mention in requirement
while(i<pred_days):
    
    if(len(temp_input)>time_step):
        
        x_input=np.array(temp_input[1:])
        #print("{} day input {}".format(i,x_input))
        x_input = x_input.reshape(1,-1)
        x_input = x_input.reshape((1, n_steps, 1))
        
        yhat = model.predict(x_input, verbose=0)
        #print("{} day output {}".format(i,yhat))
        temp_input.extend(yhat[0].tolist())
        temp_input=temp_input[1:]
        #print(temp_input)
       
        lst_output.extend(yhat.tolist())
        i=i+1
        
    else:
        
        x_input = x_input.reshape((1, n_steps,1))
        yhat = model.predict(x_input, verbose=0)
        temp_input.extend(yhat[0].tolist())
        
        lst_output.extend(yhat.tolist())
        i=i+1

predictions = scaler.inverse_transform(lst_output)
print("Output of predicted next days: \n", predictions)
```
